# Balanceo de datos

```python
# Las clases que se usaran seran las siguientes.
clases = dict(clase)
for clase, valor in clase.items():
    if valor != 1 and valor != 254:
        lista_indices = df[df["Education"]== clase].sample(n=(valor - 254), random_state=1).index
        df = df.drop(labels=lista_indices, axis=0)
    elif valor == 254:
        pass
    else:
        lista_indices = df[df["Education"]== clase].index
        df = df.drop(labels=lista_indices, axis=0)
```

# Normalizando.

```python
normalizacion = lambda valor, media, std: (valor - media)/ std

for variable in df.columns:
    if variable != "Education":
        std = df[variable].std()
        media = df[variable].mean()
        df[variable] = normalizacion(df[variable], media, std)
"""
Reseteamos los indices para poder distinguir los datos, para
cuando se haga  los sub-datasets. para cada split de los datos con 
el algoritmo de k-fold.
"""
df = df.sample(frac=1)
df = df.reset_index(drop=True)
df
```

```python
df.to_csv("marketing_campaign_normalizado.csv", index=False)
```

```python
# Algoritmo de K-fols.

def k_fold(datos, k):
    df = datos
    N = datos.shape[0]
    min_index = 0
    folds = {}
    for index in range(1, (k + 1)):
        if (index) == k:
            max_index = (int(N/k)*(index)) + (N%k) -1
        else:
            max_index = (int(N/k)*(index))
        testeo = df.iloc[min_index:max_index]
        entrenamiento = df.drop(labels=testeo.index, axis=0)
        folds["split" + str(index)] = {"entrenamiento": entrenamiento, "prueba": testeo}
        min_index = max_index
    return folds
```

```python
folds = k_fold(df, 5)
df
```

```python
# se crean los archivos para cada split(entrenamiento y testeo).
for split in folds.keys():
    folds[split]["entrenamiento"].to_csv("marketing_campaign_normalizado_"+ split +"_entrenamiento.csv")
    folds[split]["prueba"].to_csv("marketing_campaign_normalizado_"+ split +"_prueba.csv")
```

# Implementando algoritmo KNN.

```python
from knn import KNN
split1_entrenamiento = folds["split1"]["entrenamiento"].reset_index()
split1_prueba = folds["split1"]["prueba"].reset_index()
split2_entrenamiento = folds["split2"]["entrenamiento"].reset_index()
split2_prueba = folds["split2"]["prueba"].reset_index()
split3_entrenamiento = folds["split3"]["entrenamiento"].reset_index()
split3_prueba = folds["split3"]["prueba"].reset_index()
split4_entrenamiento = folds["split4"]["entrenamiento"].reset_index()
split4_prueba = folds["split4"]["prueba"].reset_index()
split5_entrenamiento = folds["split5"]["entrenamiento"].reset_index()
split5_prueba = folds["split5"]["prueba"].reset_index()
```

```python
def ejecutar_knn(k):
    a = KNN("Education", split1_entrenamiento, split1_prueba, k)
    print("Acierto split 1:", a)
    a.entrenamiento = split2_entrenamiento
    a.prueba = split2_prueba
    print("Acierto split 2:", a)
    a.entrenamiento = split3_entrenamiento
    a.prueba = split3_prueba
    print("Acierto split 3:", a)
    a.entrenamiento = split4_entrenamiento
    a.prueba = split4_prueba
    print("Acierto split 4:", a)
    a.entrenamiento = split5_entrenamiento
    a.prueba = split5_prueba
    print("Acierto split 5:", a)
```

```python
# Se ejecuta el algoritmo de KNN con k = 3, 5, 7.
ejecutar_knn(7)
```
