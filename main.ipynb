# Preprocesado de datos


```python
# se evalua si se tiene valores nulos en df
df.isna().sum()
```

```python
# Se eliminan los datos nulos, y luego verificamos.
df = df.dropna()
df.isna().sum()
```

```python
df.describe()
```

# Filtrar por id.

```python
# Se hace esto para evitar la duplicaciÃ³n de consumidores.
df = df.drop_duplicates(subset=["ID"])
df
```

# Identificando las clase.

```python
clase = df["Education"].value_counts()
clase
```

```python
# vemos todas los nombres de las columnas.
df.columns
```

# Balanceo de datos

```python
# Las clases que se usaran seran las siguientes.
clases = dict(clase)
for clase, valor in clase.items():
    if valor != 1 and valor != 54:
        lista_indices = df[df["Education"]== clase].sample(n=(valor - 54), random_state=1).index
        df = df.drop(labels=lista_indices, axis=0)
    elif valor == 54:
        pass
    else:
        lista_indices = df[df["Education"]== clase].index
        df = df.drop(labels=lista_indices, axis=0)
```

```python
df
```

# SelecciÃ³n de variables y clase

```python
variables = [
    'Year_Birth',
    'Education',
    'Income',
    'Kidhome',
    'Teenhome',
]
for variable in df.columns:
    if not (variable in variables):
        df = df.drop(columns=[variable])
df
```

# Tranformar datos categoricos.

```python
# primero se visualiza los valores de la variables categoricas.
print(df["Education"].unique())
```

```python
"""
Se daran los siquientes valores numericos a los valores de las variables categoricas.
_________
Education:
'Basic' = 1
'2n Cycle' = 2
'Graduation' = 3
'Master' = 4
'PhD' = 5
"""
df.replace(to_replace=['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD'], value=[1, 2, 3, 4, 5], inplace=True)
df
```

# Normalizando.

```python
normalizacion = lambda valor, media, std: (valor - media)/ std

for variable in df.columns:
    if variable != "Education":
        std = df[variable].std()
        media = df[variable].mean()
        df[variable] = normalizacion(df[variable], media, std)
df.reset_index()
```

```python
df.to_csv("marketing_campaign_normalizado.csv", index=False)
```

```python
# Algoritmo de K-fols.

def k_fold(datos, k):
    df = datos
    N = datos.shape[0]
    min_index = 0
    folds = {}
    for index in range(1, (k + 1)):
        if (index) == k:
            max_index = (int(N/k)*(index)) + (N%k) -1
        else:
            max_index = (int(N/k)*(index))
        print(max_index)
        testeo = df.iloc[min_index:max_index]
        entrenamiento = df.drop(labels=testeo.index, axis=0)
        folds["split" + str(index)] = {"entrenamiento": entrenamiento, "testeo": testeo}
        min_index = max_index
    return folds
```

```python
# se crean los archivos para cada split(entrenamiento y testeo).
for split in df.keys():
    df[split]["entrenamiento"].to_csv("marketing_campaign_normalizado_"+ split +"_entrenamiento.csv", index=True)
    df[split]["testeo"].to_csv("marketing_campaign_normalizado_"+ split +"_testeo.csv", index=True)
```
