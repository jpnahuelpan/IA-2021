# Preprocesado de datos

# Identificando las clase.

```python
clases = df["NumWebPurchases"].value_counts()
clases
```

# Balanceo de datos

```python
# Las clases que se usaran seran las siguientes.
clases = dict(clases)
for clase, valor in clases.items():
    if valor != 1 and valor != 43:
        lista_indices = df[df["NumWebPurchases"]== clase].sample(n=(valor - 43), random_state=1).index
        df = df.drop(labels=lista_indices, axis=0)
    elif valor == 43:
        pass
    else:
        lista_indices = df[df["NumWebPurchases"]== clase].index
        df = df.drop(labels=lista_indices, axis=0)
```

```python
df
```

# Selección de variables y clase

```python
productos = [
    "MntWines",
    "MntFruits",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "MntGoldProds",
    "NumWebPurchases",
]
for variable in df.columns:
    if not (variable in productos):
        df = df.drop(columns=[variable])
df
```

# Normalizando.

```python
normalizacion = lambda valor, media, std: (valor - media)/ std

for variable in df.columns:
    if variable != "NumWebPurchases":
        std = df[variable].std()
        media = df[variable].mean()
        df[variable] = normalizacion(df[variable], media, std)
df
```

```python
df.to_csv("marketing_campaign_normalizado.csv", index=False)
```

```python
def k_fold(datos, k):
    df = datos
    N = datos.shape[0]
    print(N)
    min_index = 0
    folds = {}
    for index in range(0, k):
        if (index + 1) == k:
            max_index = (int(N/k)*(index + 1)) + (N%k) -1
        else:
            max_index = (int(N/k)*(index + 1))
        print(max_index)
        folds[str(index)] = df.iloc[min_index:max_index]
        min_index = max_index
    return folds
```
